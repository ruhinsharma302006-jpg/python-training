## Q5: Mini Practice
Given a list of numbers, count how many are positive and how many are negative.

```python
# TODO: Count positive and negative numbers
list1=[1,2,3,-7,-3,-8,4,7,-1]
positive = 0
negative = 0

for num in list1:
    if num > 0:
        positive += 1
    elif num < 0:
        negative += 1

print("Positive numbers:", positive)
print("Negative numbers:", negative)
```
